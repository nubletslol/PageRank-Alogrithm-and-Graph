```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np 
a=nx.DiGraph()

node = ['u1','u2','u3','u4','u5','u6','u7','u8','u9']
a.add_nodes_from(node)

a.add_edges_from ([('u1','u2'),('u1','u3'),('u1','u6'),('u1','u8'),
                  ('u2','u1'),('u2','u4'),
                  ('u3','u2'),('u3','u4'),('u3','u5'),
                  ('u4','u9'),
                  ('u5','u6'),
                  ('u6','u1'),('u6','u7'),('u6','u9'),
                  ('u7','u6'),('u7','u2'),('u7','u8'),
                  ('u8','u4'),
                  ('u9','u8'),('u9','u5'),('u9','u2')])
i = 0
for name in node:
    i = i+1
    print(["%d:%s = %s"% (i,name,str(len(a.out_edges(name))))])
    
nx.draw_circular(a,node_color='',with_labels = True)
plt.show()

nx.pagerank(a,max_iter=100)
```
